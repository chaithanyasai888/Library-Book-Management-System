```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessgeBox

window=Tk()
window.title('A Simple Crud Application!')
window.geometry('1400x500')

def Database():
    global conn,cursor
    conn=sqlite3.connect('Library.db')
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS store(authorname TEXT, bookname TEXT,publishedyear INTEGER,isbn INTEGER UNIQUE)')
    
def Add_new_book():
    if str(authorname.get(1.0,'end-1c'))=="" or str(bookname.get(1.0,'end-1c'))=="" or str(publishedyear.get(1.0,'end-1c'))=="" or str(isbn.get(1.0,'end-1c'))=="":
        txt_result.config(text="Please fill in all the requried feilds",fg="red")
    
    else:
        Database()
        cursor.execute('INSERT INTO store(authorname,bookname,publishedyear,isbn) values(?,?,?,?)',
                       (authorname.get(1.0,'end-1c'),bookname.get(1.0,'end-1c'),publishedyear.get(1.0,'end-1c'),
                        isbn.get(1.0,'end-1c')))
        conn.commit()
        authorname.delete("1.0","end")
        bookname.delete("1.0","end")
        publishedyear.delete("1.0","end")
        isbn.delete("1.0","end")
        cursor.close()
        conn.close()
        txt_result.config(text="Successfully book details added in library ",fg="green")
        Display_function()
        
def Display_function():
    Database()
    tree.delete(*tree.get_children())
    cursor.execute("select * from store order by bookname ASC")
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
    conn.close()
    
        
def Display_all_records():
    Display_function()
    txt_result.config(text="! See all books available in library !",fg="green")
    
        
def Issue_a_book():
    Database()
    if isbn.get(1.0,'end-1c')=="":
         txt_result.config(text="Please enter ISBN Number to Issue",fg="red")
    else:
        a=str(isbn.get(1.0,'end-1c'))
        isbn.delete("1.0","end")
        l.append(a)
        for i in l:
            temp="DELETE FROM store where isbn = "+i
            cursor.execute(temp)   
        txt_result.config(text="Successfully Issued!",fg="green")
      
        tree.delete(*tree.get_children())
        cursor.execute("SELECT * FROM store ORDER BY bookname ASC")
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
        conn.close()
        
        
def Search_a_book():
    global temp
    
    def display():
        bookname.delete("1.0","end")
        authorname.delete("1.0","end")
        publishedyear.delete("1.0","end")
        isbn.delete("1.0","end")
        tree.delete(*tree.get_children())
        Database()
        cursor.execute(temp)
        fetch=cursor.fetchall()
        if fetch:
            for data in fetch:
                tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
            txt_result.config(text="Successfully fetched data!",fg="green")
        else:
            txt_result.config(text="Book details not founded. Plese give valid data",fg="red")
        conn.close()    
   
    if str(bookname.get(1.0, 'end-1c'))!="":
        temp1="'"+str(bookname.get(1.0, 'end-1c'))+"'"
        temp="SELECT * FROM store where bookname = "+temp1
        display()
        
    elif str(authorname.get(1.0, 'end-1c'))!="":
        temp1="'"+str(authorname.get(1.0, 'end-1c'))+"'"
        temp="SELECT * FROM store where authorname = "+temp1
        display()
            
    elif str(publishedyear.get(1.0, 'end-1c'))!="":
        temp1=str(publishedyear.get(1.0, 'end-1c'))
        temp="SELECT * FROM store where publishedyear = "+temp1
        display()
        
    elif str(isbn.get(1.0, 'end-1c'))!="":
        temp1=str(isbn.get(1.0, 'end-1c'))
        temp="SELECT * FROM store where isbn = "+temp1
        display()
    else:
        txt_result.config(text="Please fill atleast one field to search book details!",fg="red")
        
def Delete_a_book():
    if bookname.get(1.0,"end-1c")=="":
        txt_result.config(text="Please enter bookname!",fg="red")   
    else:
        temp1="'"+str(bookname.get(1.0,'end-1c'))+"'"
        temp="DELETE FROM store WHERE bookname = "+temp1
        Database()
        cursor.execute(temp)
        conn.commit()
        cursor.close()
        txt_result.config(text="Sucessfully deleted book data!",fg="green")
        Display_function()
        
def Exit():
    result=tkMessgeBox.askquestion("Do you want to exit? (y/n)",icon="warning")
    if result =="yes":
        window.destroy()
        exit()
        


Top=Frame(window,width=750,height=50,bd=8,relief='raise')
Top.pack(side=TOP)
Left=Frame(window,width=700,height=500,bd=8,relief='raise')
Left.pack(side=LEFT)
Right=Frame(window,width=800,height=900,bd=8,relief='raise')
Right.pack(side=RIGHT)

Forms=Frame(Left,width=700,height=500)
Forms.pack(side=TOP)

Buttons=Frame(Left,width=300,height=100,bd=10,relief='raise')
Buttons.pack(side=BOTTOM)


txt_title=Label(Top,width=700,font=('arial',24),text=" *** Library Book Management System ***")
txt_title.pack()

txt_authorname=Label(Forms,text="Author Name:",font=('arial',16),bd=15)
txt_authorname.grid(row=0,stick='e')
txt_bookname=Label(Forms,text="Book Name:",font=('arial',16),bd=15)
txt_bookname.grid(row=1,stick='e')
txt_publishedyear=Label(Forms,text="Published year:",font=('arial',16),bd=15)
txt_publishedyear.grid(row=2,stick='e')
txt_isbn=Label(Forms,text="ISBN:",font=('arial',16),bd=15)
txt_isbn.grid(row=3,stick='e')

txt_result=Label(Buttons)
txt_result.pack(side=TOP)


authorname=Text(Forms,height=2.5,width=30)
authorname.grid(row=0,column=1)
bookname=Text(Forms,height=2.5,width=30)
bookname.grid(row=1,column=1)
publishedyear=Text(Forms,height=2.5,width=30)
publishedyear.grid(row=2,column=1)
isbn=Text(Forms,height=2.5,width=30)
isbn.grid(row=3,column=1)


btn_update=Button(Buttons,width=15,text="Display Library",command=Display_all_records)
btn_update.pack(side=LEFT)
btn_read=Button(Buttons,width=15,text="Search a Book",command=Search_a_book)
btn_read.pack(side=LEFT)
btn_create=Button(Buttons,width=15,text="Add New Book",command=Add_new_book)
btn_create.pack(side=LEFT)
global l
l=[]
btn_delete=Button(Buttons,width=15,text="Issue a Book",command=Issue_a_book)
btn_delete.pack(side=LEFT)
btn_exit=Button(Buttons,width=15,text="Delete a Book",command=Delete_a_book)
btn_exit.pack(side=LEFT)
btn_exit=Button(Buttons,width=10,text="Exit",command=Exit)
btn_exit.pack(side=LEFT)

tree=ttk.Treeview(Right,columns=("Author Name","Book Name","Published year","ISBN"))

scrollbary=Scrollbar(Right,orient=VERTICAL)
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)

scrollbarx=Scrollbar(Right,orient=HORIZONTAL)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)

tree.heading("Author Name",text="Author Name",anchor=W)
tree.heading("Book Name",text="Book Name",anchor=W)
tree.heading("Published year",text="Published Year",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)

tree.column("#0",minwidth=0,width=0)
tree.column("#1",minwidth=0,width=130)
tree.column("#2",minwidth=0,width=135)
tree.column("#3",minwidth=0,width=140)
tree.pack()

window.mainloop()
```
